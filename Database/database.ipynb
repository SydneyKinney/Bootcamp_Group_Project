```python
# Display the list of columns
for col in data.columns:
    print(col)
```

```python
# Drop unnessasery columns
data_dropped_columns = data.drop(columns=["App Id", "Minimum Installs", "Free", "Currency", "Size", "Minimum Android", "Developer Id", "Developer Website", "Developer Email", 
"Released", "Last Updated", "Content Rating", "Privacy Policy", "In App Purchases", "Editors Choice", "Scraped Time", "Installs"])
data_dropped_columns.head()
```

```python
# Drop columns with null values
data_dropped_nulls = data_dropped_columns.dropna()
data_dropped_nulls
```

```python
# Check if new DataFrame still has null values
data_dropped_nulls.notnull().sum()
```

```python
# Replace spaces with underscores, change all letters to lowercase, and changer ad supported to numeric data
data_dropped_nulls.columns = [i.replace(' ', '_') for i in data_dropped_nulls]
clean_google_playstore_df = data_dropped_nulls.rename(columns=str.lower)
clean_google_playstore_df["ad_supported"] = clean_google_playstore_df["ad_supported"].astype(int)
clean_google_playstore_df
```

```python
clean_google_playstore_df.to_csv('Resources/clean_google_playstore.csv', index=False)
```

```python
# Create category id dataframe and csv
category_ids_df = pd.DataFrame(clean_google_playstore_df["category"].unique())
category_ids_df.insert(1, "category_id", range(1, 49))
category_ids_df.columns = ["category", "category_id"]
category_ids_df["category_id"] = category_ids_df["category_id"].astype(str).str.zfill(3)
category_ids_df.to_csv("Resources/category_id.csv", index=False)
category_ids_df
```

```python
# Create app install csv
app_install_df = clean_google_playstore_df.drop(columns=["category", "rating", "rating_count"])
app_install_df.to_csv("Resources/app_install_df.csv")
app_install_df.head()
```

```python
# Create google app info csv
google_app_info_df = clean_google_playstore_df.drop(columns=["maximum_installs", "ad_supported"])
google_app_info_df.to_csv("Resources/google_app_info.csv")
google_app_info_df.head()
```

```python
# Import dependencies to connect to database
import sqlalchemy 
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy_utils import database_exists, create_database
from local_settings import postgresql as settings
```

```python
# Create function for get engine
def get_engine(user, passwd, host, port, db):
    url = f"postgresql://{user}:{passwd}@{host}:{port}/{db}"
    engine = create_engine(url, pool_size=50, echo=False)
    return engine
```

```python
# Create engine
engine = get_engine(settings['pguser'],
                    settings['pgpasswd'],
                    settings['pghost'],
                    settings['pgport'],
                    settings['pgdb'])
```

```python
# Function to create engine
def get_engine_from_settings():
    keys = ['pguser', 'pgpasswd', 'pghost', 'pgport', 'pgdb']
    if not all(key in keys for key in settings.keys()):
        raise Exception('Bad config file')

    return get_engine(settings['pguser'],
                    settings['pgpasswd'],
                    settings['pghost'],
                    settings['pgport'],
                    settings['pgdb'])
```

```python
# Function to create session
def get_session():
    engine = get_engine_from_settings()
    session = sessionmaker(bind=engine)()
    return session
```
